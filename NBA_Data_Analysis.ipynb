```python
#Create the function to plot any data

def myviz(data, player = Players):
    col = {"KobeBryant":"Black","JoeJohnson":"Blue","LeBronJames":"Green","CarmeloAnthony":"Black","DwightHoward":"Magenta","ChrisBosh":"Red","ChrisPaul":"Blue","KevinDurant":"Green","DerrickRose":"Red","DwayneWade":"Magenta"}
    shape = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"^","ChrisBosh":"o","ChrisPaul":"s","KevinDurant":"D","DerrickRose":"^","DwayneWade":"D"}
    for name in player:
        plt.plot(data[Pdict[name]], c=col[name], ls='--', marker=shape[name], ms=7, label = name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```

```python
#Plot the free throws per game
myviz(FreeThrowAttempts/Games)
```

```python
#Accuracy of Free throws

myviz(FreeThrows/FreeThrowAttempts)
```

```python
#Playing Style of players excluding free throws

#We need to calculate points after subtractiong the free throws
#Thus, (Points - FreeThrows*1)/FieldGoals will give the true playing style of each player

myviz((Points - FreeThrows)/FieldGoals)
```

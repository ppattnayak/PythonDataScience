```python
#We need to filter out data for only 5 genres and 6 studios.
sel_genre = ["action", "comedy", "adventure", "animation","drama"]
sel_studio = ["Buena Vista Studios","Sony","Universal","WB","Paramount Pictures", "Fox"]
movies2 = movies[movies.Genre.isin(sel_genre)]
movies3 = movies2[movies2.Studio.isin(sel_studio)]
movies3.head()
```

```python
#Create the required plot by first creating a box plot and then adding a stripplot on top of it.
f, ax = plt.subplots(1,1, figsize=(8,8), sharex=True, sharey=True)
plot1 = sns.boxplot(data=movies3, x='Genre',y='GrossUSPercent',color = 'LightGray')
plot1a =sns.stripplot(data=movies3, x='Genre',y='GrossUSPercent', hue = 'Studio', size =10, jitter = True)
ax.axes.set_title("Domestic Gross % by Genre", fontsize=30)
ax.set_xlabel("Genre", fontsize = 20,)
ax.set_ylabel("Gross % US", fontsize = 20)
ax.legend(bbox_to_anchor = (1,1))
plt.show()
```
